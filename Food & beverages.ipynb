## 1. Demographic Insights

```python
df.Gender.unique()
```

```python
df.Gender.value_counts()
```

```python
#### Who prefers energy drink more? (male/female/non-binary?)
prefer_energy_drink = pd.pivot_table(
    df,
    index=["Consume_frequency"],
    columns=["Gender"],
    values=["Respondent_ID"],
    aggfunc="count"
).reset_index()

# Flatten the columns to remove the multiindex
prefer_energy_drink.columns = [col[0] if col[1] == '' else f"{col[0]}_{col[1]}" for col in prefer_energy_drink.columns]

# Reset the index to have a clean DataFrame
prefer_energy_drink = prefer_energy_drink.reset_index(drop=True)
```

```python
prefer_energy_drink
```

```python
prefer_energy_drink["Female"]=round(prefer_energy_drink['Respondent_ID_Female']/prefer_energy_drink['Respondent_ID_Female'].sum()*100,2)
prefer_energy_drink["Male"]=round(prefer_energy_drink['Respondent_ID_Male']/prefer_energy_drink['Respondent_ID_Male'].sum()*100,2)
prefer_energy_drink["Non_binary"]=round(prefer_energy_drink['Respondent_ID_Non-binary']/prefer_energy_drink['Respondent_ID_Non-binary'].sum()*100,2)
```

```python
prefer_energy_drink
```

```python
##Which age group prefers energy drinks more?
prefer_energy_drink_age = pd.pivot_table(
    df,
    index=["Consume_frequency"],
    columns=["Age"],
    values=["Respondent_ID"],
    aggfunc="count"
).reset_index()

# Flatten the columns to remove the multiindex
prefer_energy_drink_age.columns = [col[0] if col[1] == '' else f"{col[0]}_{col[1]}" for col in prefer_energy_drink_age.columns]

# Reset the index to have a clean DataFrame
prefer_energy_drink_age = prefer_energy_drink_age.reset_index(drop=True)
```

```python
df.columns
```

```python
prefer_energy_drink_age
```

```python
col=prefer_energy_drink_age.columns
col
```

```python
for i in col[1:]:
    
    b="share"
    col_name=str(i)+"_"+str(b)
    prefer_energy_drink_age[col_name]=round(prefer_energy_drink_age[i]/prefer_energy_drink_age[i].sum()*100,2)
    
```
